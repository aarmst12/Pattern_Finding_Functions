```python
'''
Define simple function PatternCount to count occurrences of string 'pat' within string 'txt'.
Slide pat along txt and evaluate equivalency at each position.
'''

def PatternCount(txt, pat):
    pattern_count = 0
    i = 0
    while i <= (len(txt) - len(pat)):
        if txt[i : (i + len(pat))] == pat:
            pattern_count += 1
        i += 1
    return pattern_count

text = input('Enter text: ')
pattern = input('Enter pattern: ')
print(PatternCount(text, pattern))
```

```python
'''
Define function MostFreqKmer which identifies the most most frequent k-mer in text.  Slide window of length k 
along text building up frequency counter. At conclusion determine which k-mer in frequency dictionary has the greatest 
value and return both k-mer and value.
'''

def MostFreqKmer(txt, k):
    j = 0
    freq = {}
    value_max = key_max = 0
    while j <= (len(txt) - k):
        pattern = txt[j : (j + k)]
        if (pattern in freq):
            freq[pattern] += 1
        else:
            freq[pattern] = 1
        j += 1
    for key, value in freq.items():
        if value > value_max:
            value_max = value
            key_max = key
    return (key_max, value_max)

text = input("Enter the text: ")
kmer = int(input("Enter k: "))
freq_kmer, kmer_count = MostFreqKmer(text, kmer)
print(freq_kmer, kmer_count)
```

```python
'''
Define function ReverseComplement to calculate reverse complement of DNA sequence.
'''

def ReverseComplement(txt):
    complement_lookup = {'A':'T', 'T':'A', 'G':'C','C':'G'}
    rev_comp = ''
    txt = txt.upper()  # account for possible user entry of lower case sequence
    i = 1
    while i <= len(txt):
        rev_comp = rev_comp + complement_lookup[txt[-i]]
        i += 1
    return (rev_comp)

text = input('Enter text: ')
print('Reverse complement is: {0}'.format(ReverseComplement(text)))
```

```python
'''
Define function PatternSites, similar to PatternCount function above except rather than enumerating occurences of 
pattern within text, lists the positions within text at which pattern occurs.
'''

def PatternSites(txt, pat):
    i = 0
    site = ''
    while i <= (len(txt) - len(pat)):
        if txt[i : (i + len(pat))] == pat:
            site = site + str(i) + " "
        i += 1
    return site
text = input("Enter text: ")
pattern = input("Enter pattern: ")
print(PatternSites(text,pattern))
```

```python
'''
Define function ClumpFinder which identifies k-mers participating in (L,t)-clumps - that is a k-mer which occurs
t or more times within a sub-sequence of length L within a possibly longer sequence text. 

Algorithm specifies starting at the beginning of text to slide a window of length k across sequence of length L 
enumerating all k-mers within L.  Evaluate which k-mers are observed t or more times, and then increment 
start position of L. L slides across text.
'''

def ClumpFinder(txt, k, L, t):
    i = 0
    clump = []
    while i <= (len(txt) - L):             #Slide L over text
        j = 0
        freq = {}
        while j <= (L - k):                 #Slide k over L
            pattern = txt[i + j : (i + j + k)]
            if (pattern in freq):
                freq[pattern] += 1
            else:
                freq[pattern] = 1
            j += 1
        i += 1
        for key, value in freq.items():     #After sliding k across L record which k-mers occur t or more times
            if value >= t:
                if key not in clump:
                    clump.append(key)
    return (clump)

text = input("Enter the text: ")
kmer = int(input("Enter k: "))
window = int(input("Enter L: "))
clump_size = int(input("Enter t: "))
clump = ClumpFinder(text, kmer, window, clump_size)
print (' '.join(x for x in clump))
```

```python
'''
Alternative algorithm for the clump finding problem.  Rather than sliding k across L at each starting position of
L, slide k across L only once at the start of text.  After L is incremented, drop the k-mer that is lost from the 
start of L and add the new k-mer formed at the end of L to the existing frequency counter, and re-evaluate which
k-mers are present t or more times.  
'''

def ClumpFinder(txt, k, L, t):
    clump = []
    freq = {}
    j=0
    while j <= (L - k):             # scan of window of length k across sequence of length L at the start of text
        pattern = txt[j : (j + k)]
        if (j == 0):                #Note first k-mer in L
            first_pat = pattern
        if (pattern in freq):
            freq[pattern] += 1
        else:
            freq[pattern] = 1
        j += 1
    for key, value in freq.items():  #After sliding k across L record which k-mers occur t or more times
            if value >= t:
                if key not in clump:
                    clump.append(key)
        
    freq[first_pat] -= 1             #Decrement frequency count of k-mer at start of sequence L prior to 
    i = 1                            #incrementing L start position
    while i <= (len(txt) - L):       #Scan L over text adding one k-mer and dropping one k-mer from the frequency 
        first_pat = txt[i : (i+k)]   #count at each new position
        last_pat = txt[i + L - k: i + L]
        if (last_pat in freq):
            freq[last_pat] += 1
        else:
            freq[last_pat] = 1

        for key, value in freq.items():
            if value >= t:
                if key not in clump:
                    clump.append(key)
        freq[first_pat] -= 1
        i += 1
    return (clump)

# provide E. coli genome as text
#with open('./E_coli.txt', 'r') as e_coli:
#    text = e_coli.read()

text = input('Enter text: ')
kmer = int(input('Enter k: '))
window = int(input('Enter L: '))
clump_size = int(input('Enter t: '))
clump = ClumpFinder(text, kmer, window, clump_size)
print(' '.join(x for x in clump))
```

```python
'''
To speed calculation can encode a k-mer as a number and record frequencies in an array indexed by these encodings.
For example, consider all possible 3-mers: AAA = 0, AAC = 1, AAG = 2, AAT = 3, ACA = 4, etc.  

Define function PatternToNumber which encodes k-mer pattern
Define function ComputeFrequencies which calculates the encoded frequency array for all k-mers within text
'''

def PatternToNumber (pattern):
    k = len(pattern)
    let_val = {'A':0, 'C':1, 'G':2,'T':3}
    value = 0
    for i in range (1,k+1):
        value += (let_val[pattern[-i]] * (4**(i-1)))
    return value

def ComputeFrequencies (text, k):
    freq_array = np.zeros(4 ** k, dtype=int)
    for i in range((len(text) - k + 1)):
        pattern = text[i:i+k]
        j = PatternToNumber(pattern)
        freq_array[j] += 1
    return(freq_array)

text = input('Enter text: ')
k = int(input('Enter k: '))
freq = ComputeFrequencies(text,k)
print(*freq)
```

```python
'''
Recursive version of PatternToNumber
'''

def PatternToNumber (pattern):
    let_val = {'A':0, 'C':1, 'G':2,'T':3}
    if not pattern:
        return 0
    else:
        symbol = pattern[-1]
        prefix = pattern[:-1]
        return 4 * PatternToNumber(prefix) + let_val[symbol]

pattern = input('Enter pattern: ')
print(PatternToNumber(pattern))
```

```python
'''
Define recursive function NumberToPattern to convert from numeric encoding back to pattern
'''

def NumberToPattern (index, k):
    val_let = ['A','C','G','T']
    if (k == 1):
        return val_let[index]
    prefixIndex = index // 4
    r = index % 4
    symbol = val_let[r]
    prefixPattern = NumberToPattern(prefixIndex, k-1)
    return (prefixPattern + symbol)

index = int(input('Enter index: '))
k = int(input('Enter k: '))
print(NumberToPattern(index, k))
```
